## Filtrando datasets

```python
data_combinada_201907["category"].value_counts()      ##Por ahora no ayuda mucho esta variable para filtrar
```

```python
data_combinada_201907["source_url"].value_counts()     ##Esta variable servirá para filtrar los proyectos tecnológicos
```

```python
solo_proy_tecnologicos = []

for i in data_combinada_201907["source_url"]:
    if i.__contains__("https://www.kickstarter.com/discover/categories/technology"):
        solo_proy_tecnologicos.append(i)
```

```python
len(solo_proy_tecnologicos)
```

```python
solo_proy_tecnologicos
```

```python
cat_tecnologico = list(set(solo_proy_tecnologicos))       ##Acá podemos ver los registros únicos de la categoría tecnología
cat_tecnologico                                           ##Funciona como un 'distinct'
```

```python
df_201907_filtrada = data_combinada_201907[data_combinada_201907['source_url'].isin (cat_tecnologico)]
df_201907_filtrada.head()
```

```python
df_201907_filtrada.shape                    ##Al final solo queda aproximadamente el 10%
```

```python
#exporta a csv (pesa el 10% ahora, es decir, 110 MB aprox)
df_201907_filtrada.to_csv( "df_201907_filtrada.csv", index=False, encoding='utf-8-sig')
```

```python
data_final = pd.read_csv("Data final limpia.csv",sep = ',')
data_final.head()
```

```python
images_list = data_final["photo"].unique()
images_list
```

```python
import urllib.request
from tqdm import tqdm
current_path = os.getcwd()
try: os.mkdir(current_path + "\\images\\")
except:pass

os.chdir(current_path + "\\images\\")

##descargar imágenes
count = 0
for img in tqdm(images_list):
    file_name = str(data_final["id"][count]) + ".png"
    urllib.request.urlretrieve(img, file_name)
    count += 1
```
